```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import random
import datetime

pages = set()
random.seed(datetime.datetime.now())
parse_result = urlparse('https://docs.python.org/zh-cn/3/library/urllib.parse.html')
# print(parse_result.scheme)
print(parse_result.netloc)
# print(parse_result.path)

def get_internal_links(bs, internal_url):
    internal_url = '{}://{}'.format(urlparse(internal_url).scheme, urlparse(internal_url).netloc)
    internal_links = []
    for link in bs.find_all('a', href = re.compile('^(/|.*' + internal_url + ')')):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in internal_links:
                if link.attrs['href'].startswith('/'):
                    internal_links.append(internal_url + link.attrs['href'])
                else:
                    internal_links.append(link.attrs['href'])

    return  internal_links


```

```python
bs = BeautifulSoup(urlopen('https://en.wikipedia.org/wiki/Janet_Sorg_Stoltzfus'), 'html.parser')
get_internal_links(bs, 'https://en.wikipedia.org/wiki/Janet_Sorg_Stoltzfus')
```

```python
def get_external_links(bs, external_url):
    external_links = []
    for link in bs.find_all('a', href = re.compile('^(http|www)((?!'+ urlparse(external_url).netloc +').)*$')):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in external_links:
                external_links.append(link.attrs['href'])

    return external_links

```

```python
def get_random_external_link(staring_page):
    html = urlopen(staring_page)
    bs = BeautifulSoup(html, 'html.parser')
    external_links = get_external_links(bs, staring_page)
    if len(external_links) == 0:
        print('No external links')
        domain = '{}://{}'.format(urlparse(staring_page).scheme, urlparse(staring_page).netloc)
        internal_links = get_internal_links(bs, domain)
        return get_random_external_link(internal_links[random.randint(0, len(internal_links)-1)])
    else:
        return external_links[random.randint(0, len(external_links)-1)]

def follow_external_only(starting_site):
    external_link = get_random_external_link(starting_site)
    print('random external link is: {}'.format(external_link))
    follow_external_only(external_link)
```

```python
all_external_links = set()
all_internal_links = set()

def get_all_external_links(site_url):
    html = urlopen(site_url)
    domain = '{}://{}'.format(urlparse(site_url).scheme, urlparse(site_url).netloc)
    bs = BeautifulSoup(html, 'html.parser')
    internal_links = get_internal_links(bs, domain)
    external_links = get_external_links(bs, domain)

    for link in internal_links:
        if link not in all_internal_links:
            all_internal_links.add(link)
            print(link)
            get_all_external_links(link)
    for link in external_links:
        if link not in all_external_links:
            all_external_links.add(link)
            print(link)


all_internal_links.add('http://oreilly.com')
get_all_external_links('http://oreilly.com')
```
